problema 2 - valorile pixelilor din imagini

```python
import matplotlib.pyplot as plt
from skimage import io
import os
import numpy as np

crtDir = os.getcwd()
file_path = os.path.join(crtDir, "images", "Karpaty.jpg")


img = io.imread(file_path)
image_normalized = img.astype(float) / 255.0


image_rgb = np.array(img)
image_normalized_rgb = np.array(image_normalized)
print(image_rgb[:, :, 0]) # red channel
print(image_normalized_rgb[:, :, 0]) # red channel

fig, axes = plt.subplots(1,2, figsize=(8,8))

axes[0].imshow(img)
axes[0].title.set_text("Before")
axes[0].axis('off')

axes[1].imshow(image_normalized)
axes[1].title.set_text("After")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
