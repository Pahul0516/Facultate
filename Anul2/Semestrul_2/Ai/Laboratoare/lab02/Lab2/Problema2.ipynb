```python
import matplotlib.pyplot as plt
from skimage import io, transform, filters, color
from skimage import feature
import os
```

Se dau mai multe imagini (salvate in folder-ul "data/images"). Se cere:

sa se vizualizeze una din imagini

```python
crtDir = os.getcwd()
file_path = os.path.join(crtDir, "images", "Karpaty.jpg")
img = io.imread(file_path)
plt.imshow(img)
plt.axis('off')
plt.show()
print(img.shape)
```

daca imaginile nu aceeasi dimensiune, sa se redimensioneze toate la 128 x 128 pixeli si sa se vizualizeze imaginile intr-un cadru tabelar.

```python
fig, axes = plt.subplots(3,4, figsize=(9,9))
img_folder = os.path.join(crtDir, "images")
img_files = [f for f in os.listdir(img_folder)]
axes = axes.ravel()
for i, ax in enumerate(axes):
    if i < len(img_files):
        img_path = os.path.join(img_folder, img_files[i])
        img = io.imread(img_path)
        rescale_img = transform.resize(img, (128, 128))
        ax.imshow(rescale_img, cmap='plasma')
        ax.set_title(f"Image {i}")
        ax.axis('off')



plt.tight_layout()
plt.show()
```

sa se transforme imaginile in format gray-levels si sa se vizualizeze

```python
fig, axes = plt.subplots(3,4, figsize=(9,9))
img_folder = os.path.join(crtDir, "images")
img_files = [f for f in os.listdir(img_folder)]
axes = axes.ravel()
for i, ax in enumerate(axes):
    if i < len(img_files):
        img_path = os.path.join(img_folder, img_files[i])
        img = io.imread(img_path, as_gray=True)
        rescale_img = transform.resize(img, (128, 128))
        ax.imshow(rescale_img, cmap='gray')
        ax.set_title(f"Image {i}")
        ax.axis('off')

plt.tight_layout()
plt.show()
```

sa se blureze o imagine si sa se afiseze in format "before-after"

```python
crtDir = os.getcwd()
file_path = os.path.join(crtDir, "images", "Karpaty.jpg")
img = io.imread(file_path)
blurred_img = filters.gaussian(img, sigma=6)

fig, axes = plt.subplots(1,2, figsize=(5,5))

axes[0].imshow(img)
axes[0].title.set_text("Before")
axes[0].axis('off')

axes[1].imshow(blurred_img)
axes[1].title.set_text("After")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

sa se identifice muchiile intr-o imagine si sa se afiseze in format "before-after"

```python
crtDir = os.getcwd()
file_path = os.path.join(crtDir, "images", "Karpaty.jpg")
img = io.imread(file_path)
gray_img = color.rgb2gray(img)
edges_img = feature.canny(gray_img, sigma=3)


fig, axes = plt.subplots(1,2, figsize=(5,5))

axes[0].imshow(img)
axes[0].title.set_text("Before")
axes[0].axis('off')

axes[1].imshow(edges_img, cmap='gray')
axes[1].title.set_text("After")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
